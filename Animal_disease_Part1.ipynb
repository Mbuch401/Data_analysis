```python
cows_df = cows_df.reset_index(drop=True)

# Display the newly indexed DataFrame
print("Cows DataFrame with reset index:")
print(cows_df.head())

# Verify the index is properly ordered
print("\nIndex range:")
print(f"Starts at: {cows_df.index.min()}")
print(f"Ends at: {cows_df.index.max()}")
```

```python
# Count cows with pneumonia
```

```python
pneumonia_count = len(cows_df[cows_df['Disease'] == 'pneumonia'])
print(f"Total number of cows with pneumonia: {pneumonia_count}")

# Calculate percentage
pneumonia_percentage = (pneumonia_count / len(cows_df)) * 100
print(f"Percentage of cows with pneumonia: {pneumonia_percentage:.2f}%")

# Visualize pneumonia cases
plt.figure(figsize=(10, 6))
disease_counts = cows_df['Disease'].value_counts()
plt.bar(['Pneumonia', 'Other Diseases'], 
        [disease_counts.get('pneumonia', 0), 
         len(cows_df) - disease_counts.get('pneumonia', 0)])
plt.title('Distribution of Pneumonia Cases in Cows')
plt.ylabel('Number of Cases')
plt.show()
```

```python
# Create DataFrame with only pneumonia cases
pneumonia_cows = cows_df[cows_df['Disease'] == 'pneumonia']

# Calculate temperature statistics for pneumonia cases
temp_stats = pneumonia_cows['Temperature'].describe()
print("Temperature Statistics for Cows with Pneumonia:")
print(f"Average Temperature: {temp_stats['mean']:.2f}F")
print(f"Minimum Temperature: {temp_stats['min']:.2f}F")
print(f"Maximum Temperature: {temp_stats['max']:.2f}F")
print(f"Standard Deviation: {temp_stats['std']:.2f}F")

# Create visualization
plt.figure(figsize=(12, 6))
sns.boxplot(x='Disease', y='Temperature', data=cows_df)
plt.title('Temperature Distribution in Cows with Pneumonia vs Other Diseases')
plt.xticks(rotation=45)
plt.show()

# Create a detailed temperature distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=pneumonia_cows, x='Temperature', bins=20, kde=True)
plt.title('Temperature Distribution in Pneumonia Cases')
plt.xlabel('Temperature (F)')
plt.ylabel('Number of Cases')
plt.axvline(x=temp_stats['mean'], color='r', linestyle='--', label='Mean Temperature')
plt.legend()
plt.show()
```

```python
unique_diseases = cows_df['Disease'].unique()
print("Types of Diseases in the Dataset:")
for i, disease in enumerate(unique_diseases, 1):
    print(f"{i}. {disease}")

# Count frequency of each disease
disease_counts = cows_df['Disease'].value_counts()
print("\nDisease Distribution:")
print(disease_counts)

# Visualize disease distribution
plt.figure(figsize=(12, 6))
disease_counts.plot(kind='bar')
plt.title('Distribution of Different Diseases in Cows')
plt.xlabel('Disease Types')
plt.ylabel('Number of Cases')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
#Age Group Analysis
```

```python
# Calculating key age statistics
age_stats = cows_df['Age'].describe()
print("Age Statistics for Cows:")
print(f"Maximum Age: {age_stats['max']:.1f} years")
print(f"Minimum Age: {age_stats['min']:.1f} years")
print(f"Median Age: {age_stats['50%']:.1f} years")

# Create a visual representation
plt.figure(figsize=(10, 6))
sns.boxplot(x=cows_df['Age'])
plt.title('Age Distribution of Cows')
plt.xlabel('Age (years)')
plt.show()

# Additional visualization with age distribution
plt.figure(figsize=(12, 6))
sns.histplot(data=cows_df, x='Age', bins=20)
plt.axvline(x=age_stats['50%'], color='r', linestyle='--', label='Median Age')
plt.axvline(x=age_stats['max'], color='g', linestyle='--', label='Maximum Age')
plt.axvline(x=age_stats['min'], color='b', linestyle='--', label='Minimum Age')
plt.title('Age Distribution with Min, Max, and Median')
plt.xlabel('Age (years)')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
# Add Age Class column and Maturity Classification columns
cows_df['Age Class'] = pd.cut(cows_df['Age'], 
                             bins=[0, 2, 15], 
                             labels=['1-2 years', '3-15 years'],
                             include_lowest=True)

# Display the first few rows to verify
print("DataFrame with new Age Class column:")
print(cows_df.head())

# Count the distribution of age classes
age_class_counts = cows_df['Age Class'].value_counts()
print("\nDistribution of Age Classes:")
print(age_class_counts)

cows_df['Maturity Status'] = np.where(cows_df['Age'] <= 2, 'Calves', 'Adults')

# Display the updated DataFrame
print("DataFrame with new Maturity Status column:")
print(cows_df.head())

# Get the count of each category
maturity_counts = cows_df['Maturity Status'].value_counts()
print("\nDistribution of Maturity Status:")
print(maturity_counts)
```

```python
cows_df.head(10)
```

```python
# Starting with calves
```

```python
# Count total number of calves
calves_count = len(cows_df[cows_df['Maturity Status'] == 'Calves'])
print(f"Total number of calves: {calves_count}")

# Calculate percentage of calves
calves_percentage = (calves_count / len(cows_df)) * 100
print(f"Percentage of calves in the population: {calves_percentage:.2f}%")
```

```python
# Get calves data
calves_df = cows_df[cows_df['Maturity Status'] == 'Calves']

# Get disease counts for calves
calves_diseases = calves_df['Disease'].value_counts()
print("Diseases in Calves:")
for disease, count in calves_diseases.items():
    print(f"{disease}: {count} cases")

# Calculate percentages
calves_diseases_percent = (calves_diseases / len(calves_df)) * 100
print("\nDisease Distribution in Calves (%):")
for disease, percentage in calves_diseases_percent.items():
    print(f"{disease}: {percentage:.2f}%")
```

```python
# Plotting calves diseases
plt.figure(figsize=(12, 8))
plt.pie(calves_diseases.values, 
        labels=calves_diseases.index,
        autopct='%1.1f%%',
        startangle=90)
plt.title('Distribution of Diseases in Calves', pad=20, fontsize=14)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.axis('equal')
plt.show()
```

```python
# Now to adult cows
```

```python
# Get adults data and count
adults_df = cows_df[cows_df['Maturity Status'] == 'Adults']
adults_count = len(adults_df)
print(f"Total number of adult cows: {adults_count}")

# Get disease counts for adults
adults_diseases = adults_df['Disease'].value_counts()
print("\nDiseases in Adult Cows:")
for disease, count in adults_diseases.items():
    print(f"{disease}: {count} cases")

# Calculate percentages
adults_diseases_percent = (adults_diseases / adults_count) * 100
print("\nDisease Distribution in Adult Cows (%):")
for disease, percentage in adults_diseases_percent.items():
    print(f"{disease}: {percentage:.2f}%")

# Summary statistics
print(f"\nTotal number of diseases found in adults: {len(adults_diseases)}")
print(f"Most common disease: {adults_diseases.index[0]} ({adults_diseases.values[0]} cases)")
```

```python
# Plotting adult cows diseases
plt.figure(figsize=(12, 8))
plt.pie(adults_diseases.values, 
        labels=adults_diseases.index,
        autopct='%1.1f%%',
        startangle=90)
plt.title('Distribution of Diseases in Adult Cows', pad=20, fontsize=14)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.axis('equal')
plt.show()
```

```python
# Create a grouped bar chart
plt.figure(figsize=(14, 8))
disease_by_age = pd.crosstab(cows_df['Age Class'], cows_df['Disease'])
disease_by_age.plot(kind='bar', stacked=True)
plt.title('Disease Distribution Across Age Groups', fontsize=14)
plt.xlabel('Age Class')
plt.ylabel('Number of Cases')
plt.legend(title='Diseases', bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()

# Calculate and display numerical breakdown
print("Disease Distribution by Age Class:")
print("\nAbsolute Numbers:")
print(disease_by_age)

print("\nPercentage Distribution:")
percentage_distribution = disease_by_age.div(disease_by_age.sum(axis=1), axis=0) * 100
print(percentage_distribution.round(2))
```

```python
# Overall disease prevalence
total_disease_counts = cows_df['Disease'].value_counts()
print("Overall Most Prevalent Diseases:")
for disease, count in total_disease_counts.items():
    print(f"{disease}: {count} cases ({(count/len(cows_df)*100):.2f}%)")

# Disease prevalence by age group
age_disease_dist = pd.crosstab(cows_df['Age Class'], cows_df['Disease'], normalize='index') * 100
print("\nDisease Distribution Within Each Age Group (%):")
print(age_disease_dist.round(2))

# Calculate disease prevalence ratio between age groups
disease_ratio = age_disease_dist.loc['1-2 years'] / age_disease_dist.loc['3-15 years']
print("\nDisease Prevalence Ratio (1-2 years vs 3-15 years):")
print(disease_ratio.round(2))

# Identify diseases with strong age group preferences
significant_difference = disease_ratio[abs(disease_ratio - 1) > 0.5]
print("\nDiseases with Strong Age Group Preferences:")
for disease, ratio in significant_difference.items():
    if ratio > 1:
        print(f"{disease}: {ratio:.2f}x more common in young cows")
    else:
        print(f"{disease}: {(1/ratio):.2f}x more common in adult cows")
```
